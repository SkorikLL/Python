```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

**Задание**:Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина
*   high - длинный(от 42)
*   middle - средний(от 35 до 42)
*   low - маленький(до 35)

```python
penguins = penguins.fillna(0)
penguins.head()
```

```python
penguins['beak_length'] = penguins['bill_length_mm']
penguins.head()
```

```python
penguins['beak_length'].astype('int')
```

```python
penguins.loc[penguins["beak_length"] < 35, "beak_length"] = "low" 
penguins.loc[(penguins["beak_length"] > 35) & (penguins["beak_length"] < 42), "beak_length"] = "middle" 
penguins.loc[penguins["beak_length"] > 42, "beak_length"] = "high"
```
